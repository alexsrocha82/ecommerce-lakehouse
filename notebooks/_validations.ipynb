```python
from pyspark.sql import functions as F

# check bronze orders
bronze_orders = spark.read.table("dev_bronze.orders")
total         = bronze_orders.count()
distinct_ids  = bronze_orders.select("order_id").distinct().count()

print(f"dev_bronze.orders")
print(f"  total rows    : {total:,}")
print(f"  distinct ids  : {distinct_ids:,}")
print(f"  duplicates    : {total - distinct_ids:,}")

# check bronze customers
bronze_custs  = spark.read.table("dev_bronze.customers")
total_c       = bronze_custs.count()
distinct_c    = bronze_custs.select("customer_id").distinct().count()

print(f"\ndev_bronze.customers")
print(f"  total rows    : {total_c:,}")
print(f"  distinct ids  : {distinct_c:,}")
print(f"  duplicates    : {total_c - distinct_c:,}")
```
